```python
plt.figure(figsize=(5, 5))
x = np.linspace(-10, 10, 1000)
y = np.exp(x)
plt.plot(x, y)
x1 = np.linspace(0.00001, 10, 1000)
y1 = np.log(x1)
plt.plot(x1, y1)
plt.ylim(-10, 10)
plt.xlim(-10, 10)

point = 550
g = np.gradient(y, x)
t = x*g[point] + (y[point] - x[point]*g[point])
plt.plot(x, t)
plt.plot(x[point], y[point], 'ro')

g1 = np.gradient(y1, x1)
t1 = x1*g1[point] + (y1[point] - x1[point]*g1[point])
plt.plot(x1, t1)
plt.plot(x1[point], y1[point], 'ro')

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
```

```python
def update_plot(xtpoint):
    plt.figure(figsize=(8, 8))
    x = np.linspace(-20, 20, 1000)
    y = np.exp(x)
    plt.plot(x, y, label="exp(x)")
    
    x1 = np.linspace(0.000000001, 20, 1000)
    y1 = np.log(x1)
    plt.plot(x1, y1, label="log(x)")
    
    plt.ylim(-20, 20)
    plt.xlim(-20, 20)
    
    # Calculate the tangent line at the given xtpoint for exp(x)
    xtindex = np.searchsorted(x, xtpoint)
    g = np.gradient(y, x)
    t = x * g[xtindex] + (y[xtindex] - x[xtindex] * g[xtindex])
    plt.plot(x, t, label="Tangent to exp(x)")
    plt.plot(x[xtindex], y[xtindex], 'ro')  # Mark the tangent point
    
    # Calculate the tangent line at the given xtpoint for log(x), clump xtpoint for log 
    if xtpoint < 0.000000001:
        xtpoint = 0.000000001
    xtindex1 = np.searchsorted(x1, xtpoint)
    g1 = np.gradient(y1, x1)
    t1 = x * g1[xtindex1] + (y1[xtindex1] - x1[xtindex1] * g1[xtindex1])
    plt.plot(x, t1, label="Tangent to log(x)")
    plt.plot(x1[xtindex1], y1[xtindex1], 'ro')  # Mark the tangent point
    
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
interact(update_plot, xtpoint=widgets.FloatSlider(value=2.5, min=-20, max=20, step=0.1, description='xtpoint'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample income data (in thousands of dollars)
income = np.array([25, 30, 35, 40, 50, 60, 1000])

# Log transformation of income data
log_income = np.log(income)

# Create subplots to visualize both distributions
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot the original income distribution
sns.histplot(income, kde=True, ax=ax[0], color='blue', bins=15)
ax[0].set_title('Original Income Distribution')
ax[0].set_xlabel('Income (Thousands)')
ax[0].set_ylabel('Frequency')

# Plot the log-transformed income distribution
sns.histplot(log_income, kde=True, ax=ax[1], color='green', bins=15)
ax[1].set_title('Log-Transformed Income Distribution')
ax[1].set_xlabel('Log(Income)')
ax[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
