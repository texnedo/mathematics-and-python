```python
import numpy as np
import re
import pandas as pd
from scipy.spatial import distance
```

```python
filtered_sentences = list()
with open('sentences.txt', 'r') as sentences_file:
    data = sentences_file.read()
    result = data.split('\n')
    result = [x.lower() for x in result if len(x) > 0]
    for sentence in result:
        filtered = re.split('[^a-zA-Z]', sentence)
        filtered = [x for x in filtered if len(x) > 0]
        filtered_sentences.append(filtered)
        print(filtered)
```

```python
uniq_words = set()
for data in filtered_sentences:
    for word in data:
        uniq_words.add(word)
uniq_words_index = list(uniq_words)
print(len(uniq_words))
```

```python
sentence_vectors = np.zeros((len(filtered_sentences), len(uniq_words)))
print(sentence_vectors.shape)
```

```python
vector_index = 0
for data in filtered_sentences:
    word_dict = dict()
    for word in data:
        word_dict[word] = word_dict.get(word, 0) + 1
    for i in range(0, len(uniq_words_index)):
        sentence_vectors[vector_index, i] = word_dict.get(uniq_words_index[i], 0)
    vector_index += 1
print(sentence_vectors) 
```

```python
dt = pd.DataFrame(sentence_vectors, columns=uniq_words_index)
dt
```

```python
vector_distances = list()
for i in range(0, len(sentence_vectors)):
    vector_distances.append((i, distance.cosine(sentence_vectors[0], sentence_vectors[i])))
vector_distances.sort(key=lambda x: x[1])
print(vector_distances)
```

```python
print("Closest vector indeces: {} and {}".format(vector_distances[1], vector_distances[2]))
```
