```python
import math
from matplotlib import pyplot as plt
import numpy as np
import pylab as pylab
from scipy import linalg

def explore_func(x: float):
    return math.sin(x / 5.0) * math.exp(x / 10.0) + 5.0 * math.exp(-x / 2.0)

def explore_func(x: np.array):
    return np.sin(x / 5.0) * np.exp(x / 10.0) + 5.0 * np.exp(-x / 2.0)
```

```python
print(explore_func(1))
print(explore_func(15))
```

```python
x = np.arange(1, 15, 0.1)
y = explore_func(x)
plt.plot(x, y)
```

```python
t = np.array([1,2,3,4])
explore_func(t)
```

```python
a = np.array([[3, 2, 0], [1, -1, 0], [0, 5, 1]])
b = np.array([2, 4, -1])
```

```python
x = linalg.solve(a, b)
print(x)
```

```python
np.dot(a, x) == b
```

```python
np.dot(np.transpose(a), b)
```

```python
#a * x = y
#w0 + w1*x = y
#w0 + w1*x + w2*x^2 = y
#w0 + w1*x + w2*x^2 + w3*x^3 = y
#
#w0 + w1*1 = 3.25
#w0 + w1*15 = 0.6
a_1 = np.array([[1, 1], [1, 15]])
x_1 = np.array([1, 15])
y_1 = explore_func(x_1)
```

```python
w_1 = linalg.solve(a_1, y_1)
```

```python
print(w_1)
```

```python
np.dot(a_1, w_1)
```

```python
def first_order(x: float):
    return w_1[0] + x * w_1[1]
```

```python
x = np.arange(1, 15, 0.1)
y_original = explore_func(x)
y_est_1 = first_order(x)
plt.plot(x, y_original)
plt.plot(x, y_est_1)
plt.show()
```

```python
a_2 = np.array([[1, 1, 1], [1, 8, 8*8], [1, 15, 15*15]])
x_2 = np.array([1, 8, 15])
y_2 = explore_func(x_2)
```

```python
w_2 = linalg.solve(a_2, y_2)
```

```python
print(w_2)
```

```python
np.dot(a_2, w_2)
```

```python
def second_order(x: float):
    return w_2[0] + x * w_2[1] + x * x * w_2[2]
```

```python
x = np.arange(1, 15, 0.1)
y_original = explore_func(x)
y_est_2 = second_order(x)
plt.plot(x, y_original)
plt.plot(x, y_est_2)
plt.show()
```

```python
a_3 = np.array([[1, 1, 1, 1], [1, 4, 4*4, 4*4*4], [1, 10, 10*10, 10*10*10], [1, 15, 15*15, 15*15*15]])
x_3 = np.array([1, 4, 10, 15])
y_3 = explore_func(x_3)
```

```python
w_3 = linalg.solve(a_3, y_3)
```

```python
def third_order(x: float):
    return w_3[0] + x * w_3[1] + x * x * w_3[2] + x * x * x * w_3[3]
```

```python
x = np.arange(1, 15, 0.1)
y_original = explore_func(x)
y_est_3 = third_order(x)
plt.plot(x, y_original)
plt.plot(x, y_est_3)
plt.show()
```
